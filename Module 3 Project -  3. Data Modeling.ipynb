# Modeling

For this project, I am going to take the brute force approach and try a few different supervised learning algorithms in order to see which works best. While this approach is a little time consuming, it will provide me with the best results and allow me to compare training and testing scores. Our goal is to correctly predict Alcohol users. Before I get to modeling I am going to create a model function that easily builds models and displays results so that going through each model will be smoother and more organized.

```python
X = df.drop(columns = ['Alcohol'], axis = 1) #Splitting features
y = df['Alcohol'] #Splitting target variable

def build_model(classifier, predictors, labels): #Creating function to build models and display results
    
    #Splitting the data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, random_state = 34)
    
    #Scaling the data
    scaler = MinMaxScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    #Make variables to be called outside of the function
    build_model.X_train = X_train
    build_model.X_test = X_test
    build_model.y_train = y_train
    build_model.y_test = y_test
    
    #Fit the model
    model = classifier.fit(X_train, y_train)
    score = np.mean(cross_val_score(classifier, X, y, cv = 5, scoring = 'recall'))
    
    #Create predictions
    y_hat_train = model.predict(X_train)
    y_hat_test = model.predict(X_test)
    
    #Make variables to be called outside of the function
    build_model.y_hat_train = y_hat_train
    build_model.y_hat_test = y_hat_test
    
    #Print classification report
    print('Train Classification Report\n', classification_report(y_train, y_hat_train))
    print('\n-------------------------------\n')
    print('Test Classification Report\n', classification_report(y_test, y_hat_test))
    print('\n-------------------------------\n')
    
    print('Train Accuracy:', round(accuracy_score(y_train, y_hat_train), 3))
    print('Test Accuracy:', round(accuracy_score(y_test, y_hat_test), 3))
    print('Cross Validation Recall', round(score, 3))
    print('\n-------------------------------\n')
    
    #Create a results dataframe
    results = pd.DataFrame([[str(model), round(accuracy_score(y_train, y_hat_train), 3), 
                            round(accuracy_score(y_test, y_hat_test), 3), round(score, 3)]], 
                          columns = ['Model', 'Train_Accuracy', 'Test_Accuracy', 'Cross_Val_Recall'])
    build_model.results = results
    
    #Plot confusion matrices
    fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (15, 10))
    plot_confusion_matrix(classifier, X_train, y_train, cmap = plt.cm.plasma, ax = axes[0])
    fig.suptitle('Train and Test Confusion Matrices', fontsize = 20);
    
    plot_confusion_matrix(classifier, X_test, y_test, cmap = plt.cm.plasma, ax = axes[1])
    plt.subplots_adjust(wspace = 0.4)
    
    return model
```

## Plot ROC Curve and Compare Scores

In the following cells I plot the ROC curve to determine which model is best at predicting the true labels. I will also train-test split the data again just outside of the original function above.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, random_state = 34) #Splitting again
```

```python
models = [logreg_model, mnb_model, dt_model, rf_model, brf_model, ada_model] #Listing models
model_names = ['LogReg', 'MNB', 'DT', 'RF', 'BRF', 'ADA'] #Listing model names
```

```python
def roc(models, names): #Creating function to build roc curves
    
    #Creating lists for scores
    aucs = []
    fprs = []
    tprs = []

    for i, model in enumerate(models): #Creating for loop for model functions
        try:
            y_score = model.decision_function(X_test)
            fpr, tpr, thresh = roc_curve(y_test, y_score)

        except AttributeError:
            y_score = model.predict_proba(X_test)
            fpr, tpr, thresh = roc_curve(y_test, y_score[:, 1])


        aucs.append(auc(fpr, tpr))
        fprs.append(fpr)
        tprs.append(tpr)

    plt.figure(figsize=(15, 12)) #Creating figure and colors for roc curve lines
    lw = 2
    colors = ['red', 'darkorange', 'yellow', 'green', 'darkblue', 'pink', 'teal', 'gainsboro', 'purple', 
             'aqua', 'tomato', 'tan', 'azure']

    for n in range(len(models)): #for loop plotting lines
        plt.plot(fprs[n], tprs[n], color = colors[n],
                 lw=lw, label= f'{names[n]}, AUC: {round(aucs[n], 3)}')

    plt.plot([0, 1], [0, 1], color = 'navy', lw = lw, linestyle = '--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.yticks([i/20.0 for i in range(21)])
    plt.xticks([i/20.0 for i in range(21)])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc = 'lower right')
    plt.show()
```

```python
roc(models, model_names) #Plotting ROC
```

```python
def compare_scores(models, names): #Create function for comparing scores
    
    #Create lists for scores
    train_accs = []
    test_accs = []
    train_f1s = []
    test_f1s = []

    for model in models: #For loops for model scores

        model.fit(X_train, y_train)

        yhat_train = model.predict(X_train)
        yhat_test = model.predict(X_test)

        train_acc = accuracy_score(y_train, yhat_train)
        train_accs.append(round(train_acc, 3))

        test_acc = accuracy_score(y_test, yhat_test)
        test_accs.append(round(test_acc, 3))

        train_f1 = f1_score(y_train, yhat_train)
        train_f1s.append(round(train_f1, 3))

        test_f1 = f1_score(y_test, yhat_test)
        test_f1s.append(round(test_f1, 3))

    model_scores = pd.DataFrame({'Model': names,
                                 'Training Accuracy': train_accs,
                                 'Testing Accuracy': test_accs,
                                 'Training F1 Score': train_f1s,
                                 'Testing F1 Score': test_f1s})
    
    return model_scores
```

```python
model_scores = compare_scores(models, model_names) #Putting models in comparing scores function
model_scores #Printing table comparing scores
```
