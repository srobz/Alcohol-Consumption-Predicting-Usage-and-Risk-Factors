As mentioned in the previous notebook, N-, E-, O-, A-, and CScore are all personality measurements. It is important to note here what these scores mean, as it will help us further understand these features. These personality measurements are scored on a scale; higher scores mean a presence of certain traits while lower scores mean a presence or lack of other traits.

* Neuroticism: Adjustment/emotional stability or maladjustment/neuroticism.
    * Individuals with higher scores here are more prone to irrational ideas, are less able to control impulses, and have trouble when coping with stress.
    * Individuals with lower scores are typicaly calm, relaxed, and able to face stressful situations without becoming upset.
* Extraversion: Sociability/absence of sociability (rather than the opposite).
    * Higher scores indicate sociability, preference towards large groups and gatherings, as well as traits such as being assertive, active, and talkative. 
    * Lower scores indicate introversion; reservation rather than unfriendliness, introverts are not depressed or unhappy.
* Openness to Experience: 
    * Higher scores are associated with education and intelligence but is not equivalent. Tend to lead experientially richer lives.
    * Lower scores are associated with conventional behavior and conservative outlooks. Tend to have a narrower scope and less intense interests
* Agreeableness: Associated with interpersonal tendencies.
    * Higher scores indicate altruism, sympathy, and an eagerness to help.
    * Lower scores tend to be antagonistic, egoenctric, skeptical of others, and competitive.
* Conscientiousness: Control of impulses.
    * Higher scores indicate punctuality, reliability, and persistance. They are usually well-organized and see life in terms of accomplished tasks.
    * Lower scores are typical of those that are less goal-seeking and less reliable in their moral standards.

Source: 
* http://www.csun.edu/~hcpsy002/NEO_Interpretation_Info.pdf

These continuous variables indicate a lot about an individual, and can help tell us whether or not someone is more or less likely to be a drug user. As we continue throughout this analysis, it will be helpful to keep this information in mind.

## Dummy Variables

In the following cells I will push the categorical variales into dummies.

```python
df2 = pd.get_dummies(df2) #Getting dummies
df2.info() #Checking to make sure dummies were made
```

## Check for Multicollinearity

I'm going to check for multicollinearity and remove any features with a pairwise correlation above 0.75.

```python
corr = df2.corr().abs() #Setting correlation
c = corr.unstack() #Setting c
co = c.sort_values(kind = 'quicksort') #Setting co
corr.head() #Print 5 rows of correlations
```

```python
co.loc[co >= .75] #Double checking which predictors have overly high pairwise-correlation
```

We're good to go, none of the predictors have multicollinearity.

## Feature Selection

In the following cells, I perform a simple feature selection using Statsmodel's Ordinary Least Squares (OLS) in order to determine the best features to use when modeling. I will determine this by removing any features that have p-values greater than 0.05.

```python
y = df2['Alcohol'] #Define target
X = df2.drop('Alcohol', axis = 1) #Define features

X = sm.add_constant(X) #Define the intercept

model = sm.OLS(y, X) #Fit the model
result = model.fit() #Set the result

result.summary() #Summarize the results
```

Many features had p-values above 0.05, but before I go ahead and remove said features I am going to walk through them by category:

* All of the personality score measurements have high p-values, so they will all be removed.
* Both cannabis and nicotine are above 0.05, so they will be removed.
* Those aged 55-64 and 65+ will be removed.
* E1 (Left school before 16 years), E2 (Left school at 16 years), and E4 (Left school at 18 years) will be removed.
* Canada, New Zealand, and Ireland will be removed.
* Ethnicity-Black and Ethnicity- Mixed Black/Asian will be removed.

While I can't help but wonder how some of these features would affect the data - especially the personality score measurements and the levels of education, it is best to remove them since their p-values are above 0.05.

```python
df2 = df2[['Alcohol', 'Caff', 'Age_18-24', 'Age_25-34', 'Age_35-44', 'Age_45-54', 'Gender_Female', 'Gender_Male', 
          'Education_E3', 'Education_E5', 'Education_E6', 'Education_E7', 'Education_E8', 'Education_E9', 
          'Country_Australia', 'Country_Other', 'Country_UK', 'Country_USA', 'Ethnicity_Asian', 
          'Ethnicity_Mixed-White/Asian', 'Ethnicity_Mixed-White/Black', 'Ethnicity_Other', 'Ethnicity_White']].copy()
df2.head()
```
