Things to Note:
* Age is grouped in increments rather than individually known for each entry, meaning it is a categorical feature when it would usually be continuous.
* I gave Education monikers for easier data analysis as they had longer explanations, below are the given explanations and monikers:
    * E1 = Left school before 16 years
    * E2 = Left school at 16 years
    * E3 = Left school at 17 years
    * E4 = Left school at 18 years
    * E5 = Some college or university, no certificate or degree
    * E6 = Professional certificate/diploma
    * E7 = University degree
    * E8 = Masters degree
    * E9 = Doctorate degree
* N-, E-, O-, A-, and CScore are personality measurements meaning neuroticism, extraversion, openness to experience, agreeableness, and conscientiousness; these are continuous variables.

"Impulsive" which was a measure of impulsivity and "SS" which was a measure of sensation seeking didn't have proper information to replace their coded numbers, so as much as I would like to include that information I will be dropping these columns as they can not be helpful in this analysis. I will also be changing the scores of the personality measurements to integers from strings.

```python
df = df.drop(columns = ['Impulsive', 'SS'], axis = 1) #Dropping columns that didn't have proper information
df['NScore'] = df['NScore'].astype(int) #Change from str to int
df['EScore'] = df['EScore'].astype(int) #Change from str to int
df['OScore'] = df['OScore'].astype(int) #Change from str to int
df['AScore'] = df['AScore'].astype(int) #Change from str to int
df['CScore'] = df['CScore'].astype(int) #Change from str to int
df.info() #Check that types were properly changed
```

Next I will remove any users that said they used the drug "Semer," as this was a fake drug created for the purpose of weeding out people who over-claimed.

```python
df = df[df.Semer == 'CL0'] #Setting dataframe to non-users of Semer
df.head() #Checking first 5 rows
```

Now I'm going to drop the drugs VSA (Volatile Substance Abuse), Legal Highs, and Semer, as they will not be helpful for the analysis I want to do.

```python
df = df.drop(columns = ['VSA', 'Legalh', 'Semer'], axis = 1) #Removing said drugs from dataframe
```

For this analysis, I am going to just look at whether or not each person is a user of the drug. From the original dataset, we know that CL0 through CL6 each has a specific meaning:
* CL0 = Never Used
* CL1 = Used over a Decade Ago
* CL2 = Used in Last Decade
* CL3 = Used in Last Year
* CL4 = Used in Last Month
* CL5 = Used in Last Week
* CL6 = Used in Last Day

We are going to turn this into a binary classification, if the person has CL0-CL2, they are not a user, and if they have CL3-CL6, they are a user of that drug. In the following cell I run a for loop that replaces the information with binary code.

```python
drugs = ['Alcohol', 'Amphet', 'Amyl', 'Benzos', 'Caff', 'Cannabis', 'Choc', 'Coke', 'Crack', 
          'Ecstasy', 'Heroin', 'Ketamine', 'LSD', 'Meth', 'Mushrooms', 'Nicotine'] #List of drugs in dataframe
for drug in drugs: #Creating for loop of drugs
    df[drug].replace({'CL0': 0, 'CL1': 0, 'CL2': 0, 'CL3': 1, 'CL4': 1, 'CL5': 1, 'CL6': 1}, inplace = True) #Replaced

df.head() #Checking first 5 rows
```
